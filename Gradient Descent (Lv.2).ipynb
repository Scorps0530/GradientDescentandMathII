```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random dataset
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Initialize parameters for linear regression
theta0 = np.random.randn()
theta1 = np.random.randn()

# Learning rate and number of iterations
learning_rate = 0.1
iterations = 20

# Store values of theta0 and theta1 for plotting
theta0_values = [theta0]
theta1_values = [theta1]

# Perform gradient descent
for _ in range(iterations):
    predictions = theta0 + theta1 * X
    d_theta0 = (1 / len(y)) * np.sum(predictions - y)
    d_theta1 = (1 / len(y)) * np.sum((predictions - y) * X)
    theta0 -= learning_rate * d_theta0
    theta1 -= learning_rate * d_theta1
    theta0_values.append(theta0)
    theta1_values.append(theta1)

# Plotting the dataset and regression lines
plt.figure(figsize=(20, 10))
plt.scatter(X, y, label='Data Points')

# Plot regression lines at different stages
for i in range(0, len(theta0_values), 5):
    y_pred = theta0_values[i] + theta1_values[i] * X
    plt.plot(X, y_pred, label=f'Iteration {i}', linestyle='--')

# Final regression line
y_pred = theta0_values[-1] + theta1_values[-1] * X
plt.plot(X, y_pred, 'r-', label='Final Regression Line')

plt.title('Linear Regression using Gradient Descent')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its gradient
def f(x):
    return x**2

def grad_f(x):
    return 2 * x

# Gradient descent parameters
learning_rate = 0.1
initial_x = 5
iterations = 20

# Store the values of x and f(x) for plotting
x_values = [initial_x]
y_values = [f(initial_x)]

# Perform gradient descent
x = initial_x
for _ in range(iterations):
    x = x - learning_rate * grad_f(x)
    x_values.append(x)
    y_values.append(f(x))

# Plotting the function and the gradient descent steps
x_range = np.linspace(-6, 6, 400)
y_range = f(x_range)

plt.figure(figsize=(10, 6))
plt.plot(x_range, y_range, label='f(x) = x^2')
plt.scatter(x_values, y_values, color='red')
plt.plot(x_values, y_values, color='red', linestyle='--', label='Gradient Descent Path')
plt.title('Gradient Descent Optimization')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
